```r
### Played Hour Distribution and Suscription Status Graph ###


options(repr.plot.width = 11)
played_hour_distribution_plot <- ggplot(player_data, aes(x = played_hours, fill = subscribe)) +
geom_histogram(bins = 450, color = "white") +
scale_fill_discrete(labels = c("No", "Yes")) +
coord_cartesian(xlim = c(0, 20)) +
labs(title = "Distributions of Played Hours and Subscription Status",
     x = "Hours Played",
     y = "Number of Players",
     fill = "Subscription?") +
theme(text = element_text(size = 20))

played_hour_distribution_plot
```

Outliers were not included in this graph so we can zoom in on the big picture. This plot shows that played_hours is extremely right-skewed, with most players between 0-3 total hours. Both subscribers and non-subscribers appeared in this range. At higher played hours, nearly all players are subscribers. This suggests that players with higher played hours are more likely to subscribe.

```r
### Gameplay Consistency and Subscription Status Graph ###

options(repr.plot.width = 11)
session_count <- session_data |>
count(hashed_email, name = "session_count")

consistency_data <- player_data |>
left_join(session_count, by = "hashed_email") |>
mutate(session_count = coalesce(session_count, 0))

gameplay_consistency_plot <- ggplot(consistency_data, aes(x = session_count, fill = subscribe)) +
geom_histogram(bins = 450, color = "white") +
scale_fill_discrete(labels = c("No", "Yes")) +
coord_cartesian(xlim = c(0, 50)) +
labs(title = "Gameplay Consistency and Subscription Status",
     x = "Number of Sessions",
     y = "Number of Players",
     fill = "Subscription?") +
theme(text = element_text(size = 20))

gameplay_consistency_plot
```

Again, outliers are eliminated. This plot also shows a strong right skew. Most players have 0-2 sessions, and players with more sessions are mostly subscribers. Both of the plots may indicate class imbalance between subscribed and non-subscribed players. 

4. Methods and Plan

Since the subscription variable is binary, I will be using K-NN classification to solve my question. Classification is suitable because it predicts categorical outcomes and can incorporate multiple predictors. The main assumptions are that similar players are close in feature space and that the predictors are standardized so distances are meaningful. Limitations include sensitivity to outliers and the effect of skewed distribution, and overlap in low-session values. To choose the best k, I will use cross validation and compare the accuracy across different k values. The data will be split into 70% training set and 30% testing set, so there will be enough data to train the model, and enough data to test the model.
