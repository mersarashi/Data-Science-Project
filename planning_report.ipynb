The data set "player.csv" stores information of a list of players and their data. This dataset includes 196 rows and 7 variables, with each row representing one player. The variables and their descriptions are listed below:


| Variable | Data Type | Description |
|----------|------------|------------------------------|
| experience | Character | Describes the experience level of each player |
| subscribe | Logical | Shows if the player subscribed to a game-related newsletter (TRUE or FALSE) |
| hashed_email | Character | Hashed version of player's email address |
| played_hours | Double | Indicates the number of gameplay hours recorded for the player |
| name | Character | Name of the player |
| gender | Character | Gender of the player |
| age | Double | Age of the player |

- The variable "experience has five categories: pro, veteran, amateur, regular or beginner.

- The variable "gender" is categorized into: male, female, non-binary, prefer not to say, agender, two-spirited and others.

The dataset "session_data" records how long the players play Minecraft in each session. There are 1535 rows and each row represents a session. The variable original_start_time and original_end_time are likely be recorded in milliseconds, and they are converted to improve readability. Details are listed below: 

| Variable | Data Type | Description |
|----------|------------|------------------------------|
| hashed_email | Character | Hashed version of player's email address |
| start_time | Character | The time when the session begins. Converted from the variable "original_start_time" to improve readability|
| end_time | Character | The time when the session ends. Converted fromm the variable "original_end_time" to improve readability |
| original_start_time | Double | The original start timestamp before conversion |
| original_end_time | Double | The original end timestamp before conversion |
